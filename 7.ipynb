bubble sort

```python
arr=[2,7,99,5,6,7]
def bubble(arr):
  n=len(arr)
  for i in range(n-1):
    for j in range(n-i-1):
      if arr[j]>arr[j+1]:
        temp=arr[j]
        arr[j]=arr[j+1]
        arr[j+1]=temp
  return arr
result=bubble(arr)
print(result)
```
