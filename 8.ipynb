```python
arr = [2, 1, 4, 0, 8, 6]

def selection_sort(a):
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        # Swap the found minimum element with the first element
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

result = selection_sort(arr)
print(result)
```
