```python
A="ASHISH"
for i in range len(A):
  A[0]!==A[i+1]
  print(A)
else:
  print(A[i+1],A)
```

```python
a="ashish"
counts={}

for char in a:
  if char in counts:
    counts[char] +=1
  else:
      counts[char]=1
for char,frequency in counts.items():
  if frequency>1:
    print(f"{char} {frequency}")
```
