```python
def find_sum_pair(a, target):
  n=len(a)
  for i in range(n-1):
    for j in range(i+1,n):
      if a[i]+ a[j]==target:
        return [i,j] # This return is inside the function
  return None # This return is also inside the function

a=[1,2,3,4]
target=6
result = find_sum_pair(a, target)
print(result)
```

```python
a = [1, 2, 3, 4]
target = 6
n = len(a)
result_no_func = None

for i in range(n - 1):
  for j in range(i + 1, n):
    if a[i] + a[j] == target:
      result_no_func = [i, j]
      break # Exit inner loop once a pair is found
  if result_no_func:
    break # Exit outer loop once a pair is found

print(result_no_func)
```

This code achieves the same result by directly iterating through the list and checking for the sum, storing the result in `result_no_func`.
