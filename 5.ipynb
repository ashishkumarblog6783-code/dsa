sum of first n number

```python
def sum(n):
  if(n==0):
    return 0
  return n+sum(n-1)

print(sum(5))
```

```python
arr=[5,4,3,2,1]
def sum(n):
  if(n==0):
    return arr[n]
  return arr[n]+sum(n-1)
print(sum(len(arr)-1))
```

```python
arr=[1,4,5,7]
def sum(n):
  is_odd=(arr[n]%2!=0)
  if(n==0):
    return arr[n] if is_odd else 0
  return (arr[n] if is_odd else 0) + sum(n-1)
print(sum(len(arr)-1))
```

```python
arr=[2,4,5,6]
def sum(n):
  is_even=(arr[n]%2==0)
  if(n==0):
    return arr[n] if is_even else 0
  return (arr[n] if is_even else 0)+ sum(n-1)
print(sum(len(arr)-1))
```

```python
def factorial(n):
  if(n==0):
    return 1
  else:
    return n * (n-1) # Corrected: return the product and make a recursive call

print(factorial(2)) # Added print to see the output
```
